```python
def add_temporal_context(df):
    df['time'] = pd.to_datetime(df['time'])

    hour = df['time'].dt.hour
    df['hour_sin'] = np.sin(2 * np.pi * hour / 24.0)
    df['hour_cos'] = np.cos(2 * np.pi * hour / 24.0)

    day = df['time'].dt.dayofweek
    df['day_sin'] = np.sin(2 * np.pi * day / 7.0)
    df['day_cos'] = np.cos(2 * np.pi * day / 7.0)

    return df
```

```python
behaviors = add_temporal_context(behaviors)
```

```python
def parse_impressions(impressions):
    items = impressions.split()
    data = []
    
    for item in items:
        news_id, label = item.split("-")
        data.append((news_id, int(label)))
        
    return data

# test
print(parse_impressions(row["impressions"]))
```

```python
processed_behaviour = []

for _, row in behaviors.iterrows():
    history = row["history"]
    
    if pd.isna(history):
        continue
        
    parsed = parse_impressions(row["impressions"])
    
    for news_id, label in parsed:
        processed_behaviour.append({
            "user_id": row["user_id"],
            "history": history,
            "news_id": news_id,
            "label": label
        })

df = pd.DataFrame(processed_behaviour)
print(df.head())
```

```python
#Remove empty history
df = df[df["history"].notna()]
```

```python
df["history"] = df["history"].apply(lambda x: x.split())
```

```python
# Select useful columns
news_clean = news[["news_id", "category", "subcategory", "title", "abstract"]].copy()

# Handle missing values
news_clean["category"] = news_clean["category"].fillna("unknown")
news_clean["subcategory"] = news_clean["subcategory"].fillna("unknown")
news_clean["title"] = news_clean["title"].fillna("")
news_clean["abstract"] = news_clean["abstract"].fillna("")

# Remove duplicates
news_clean = news_clean.drop_duplicates(subset="news_id")

# Reset index
news_clean = news_clean.reset_index(drop=True)

print(news_clean.head())
print("Shape:", news_clean.shape)
```

```python
behaviors_clean = df.copy()


# Remove rows with empty history (optional but recommended)
behaviors_clean = behaviors_clean[behaviors_clean["history"].notna()]


# Reset index
behaviors_clean = behaviors_clean.reset_index(drop=True)

print(behaviors_clean.head())
print("Shape:", behaviors_clean.shape)
```

```python
news_clean.to_csv("data/cleaned/news_clean.csv", index=False)
behaviors_clean.to_csv("data/cleaned/behaviors_clean.csv", index=False)
```
